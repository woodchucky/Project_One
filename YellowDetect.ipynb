```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

rockimg = mpimg.imread('../calibration_images/example_rock2.jpg')
plt.imshow(rockimg)
```

```python
def color_thresh2(img):
    color_select = np.zeros_like(img[:,:,0])
    in_rock_thresh = (img[:,:,2]<50) \
    & (img[:,:,0] > 100) & (img[:,:,1] > 100)
    color_select[in_rock_thresh] = 1
    return color_select
rockimg2 = color_thresh2(rockimg)
plt.imshow(rockimg2, cmap='gray')
```
